<h1><b>SVM practice dataset model</b></h1>

```python
import pandas as pd 
df = pd.read_csv('Cancer_Diagnostic_Dataset.csv')
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['diagnosis_n'] = le.fit_transform(df['diagnosis'])
df.head()
```

```python
x = df.drop(['diagnosis', 'id', 'Unnamed: 32','diagnosis_n'], axis=1)
y = df['diagnosis_n']
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2,)
```

```python
len(x_train), len(x_test)
```

```python
from sklearn.svm import SVC
model = SVC()
```

```python
model.fit(x_train, y_train)
```

```python
model.predict(x_test)
```

```python
model.score(x_test, y_test)
```

<h1><b>Tuning:-</h1></b><br>
Using kernel, Gamma, and regularization(C)


```python
model_01 = SVC(kernel='linear')
model_01.fit(x_train, y_train)  
model_01.score(x_test, y_test)
```

```python
model_02 = SVC(kernel='poly')
model_02.fit(x_train, y_train)  
model_02.score(x_test, y_test)
```

```python
model_03= SVC(kernel='sigmoid')
model_03.fit(x_train, y_train)  
model_03.score(x_test, y_test)
```

```python
model_04= SVC(gamma = 0.0001, kernel='rbf',C = 3.05)   #highest accuracy model 
model_04.fit(x_train, y_train)  
model_04.score(x_test, y_test)
```

```python
model_05= SVC(C = 400)
model_05.fit(x_train, y_train)  
model_05.score(x_test, y_test)
```
