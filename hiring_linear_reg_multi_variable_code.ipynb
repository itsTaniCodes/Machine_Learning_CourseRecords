Installing word2number<br>
it converts word into integer

```python
from word2number import w2n
import pandas as pd
import numpy as np

def safe_word_to_num(x):
    if pd.isna(x):  # Check for actual NaN
        return np.nan
    if isinstance(x, str) and x.lower() == 'nan':  # Check for string "nan"
        return np.nan
    try:
        return w2n.word_to_num(x)
    except:
        return np.nan  # Return NaN if conversion fails

df['experience'] = df['experience'].apply(safe_word_to_num)

```

<h>Getting Median<h>

```python
import math 
df['experience'].median()
floor_median_exp = math.floor(df['experience'].median())
floor_median_exp        
```

<h>filling up the empty spaces<h>

```python
df['experience'].fillna(floor_median_exp)
```

Adding changes to dataframe

```python
df['experience'] =df['experience'].fillna(floor_median_exp)
df
```
